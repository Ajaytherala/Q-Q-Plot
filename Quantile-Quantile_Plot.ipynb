1. Plot for determining whether a random variable is gaussian distributed or not


```python
import numpy as np
import pylab
import scipy.stats as stats
```

```python
gaussian_distribution = np.random.normal(loc = 20, scale=5, size=1000)
stats.probplot(gaussian_distribution,dist = "norm",plot = pylab)
pylab.show()
```

```python
uniform_distribution = np.random.uniform(low = 20, high=50, size=1000)
stats.probplot(uniform_distribution,dist = "norm",plot = pylab)
pylab.show()
```

1. In First case we can see that most of the points are on the line where as in the second scenario we can see that most of the points are away from the line.

2. Based upon the graph we can determine whether a random variable is gaussian distributed or not.

3. It will be hard to interpret Q-Q Plot if no.of samples are small as shown in below graph.

4. dist = "norm" is nothing but standard normal variate => Z ~ N(0,1) => Random variable that is Gaussian distributed with mean 0 and variance 1
